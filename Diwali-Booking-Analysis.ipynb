```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#%matplotlib.inline
import seaborn as sns
```

```python
df=pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']=df['Amount'].astype('int')
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

# Exploratory Data Analyis

# Gender

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

## Age

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

### From the above graph we can conclude female are the highest spender

## State

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
sales_state_amt=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,12)})
sns.barplot(x='State',y='Amount',data=sales_state_amt)
```

*We can conclude*

```python
ax=sns.countplot(x='Marital_Status',data=df,hue='Gender')
sns.set(rc={'figure.figsize':(5,6)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_marital=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set(rc={'figure.figsize':(18,12)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_marital,hue='Gender')
```

Married women are the highest spender

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(20,15)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occ=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
#sns.set(rc={'figure.figsize':(18,12)})
sns.barplot(x='Occupation',y='Amount',data=sales_occ)
```

### Product

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(20,18)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_prod=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
#sns.set(rc={'figure.figsize':(18,12)})
sns.barplot(x='Product_Category',y='Amount',data=sales_prod)
```

```python
sales_pid=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
#sns.set(rc={'figure.figsize':(18,12)})
sns.barplot(x='Product_ID',y='Orders',data=sales_pid)
```
